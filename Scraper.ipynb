```python
import numpy as np
import pandas as pd
from selenium.webdriver import Chrome
from bs4 import BeautifulSoup
import time
from selenium.webdriver.chrome.options import Options

chrome_options = Options()
chrome_options.add_argument("--window-size=1920,1080") #adjust window size of selenium driver to avoid mobile view
chrome_options.add_argument("--headless")
webdriver = "/Users/lukas/Documents/GitHub/googlemaps-scraper/chromedriver" #locate webdriver
```

```python
driver = Chrome(webdriver, options=chrome_options) #launch chrome webdriver
#launch result url
result_url_2018_2019 = "https://koeln-marathon.de/ergebnisse/#1_1A0B0C" #url for results from 2018 and 2019

driver.get(result_url_2018_2019)
driver.implicitly_wait(10)
```

```python
all_results_2019_button = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[2]/div[2]/div/div[5]/table/tbody[2]/tr/td/a[2]')
all_results_2019_button.click() #click button to view all 2019 results

time.sleep(8)


#Selenium hands the page source to BeautifulSoup
results_soup=BeautifulSoup(driver.page_source, 'lxml')
```

```python
#table [0] is 2019 results
table_2019_rows = results_soup.findChildren('table')[0].findAll("tr", {"class": "Hover LastRecordLine"})
```

```python
results_2019 = []

for index, tr in enumerate(table_2019_rows):
    runner_id = index+1
    td = tr.find_all('td')
    runner_name = driver.find_element_by_xpath(f'/html/body/div[2]/div[2]/div[2]/div[2]/div/div[5]/table/tbody[1]/tr[{runner_id}]/td[4]')
    runner_name.click()
    time.sleep(2)
    #Selenium hands the page source to Beautiful Soup
    results_detail_soup=BeautifulSoup(driver.page_source, 'lxml')
    rows_detail = results_detail_soup.findChildren('table')[2].findAll("tr", {"class": "Hover LastRecordLine"})
    row_detail = []
    for tr_detail in rows_detail[5:14]:
        td_detail = tr_detail.find_all('td')
        row_detail += [tr_detail.text for tr_detail in td_detail][1:2]
    exit_button = driver.find_element_by_xpath('/html/body/table/tbody/tr/td/div/img')
    exit_button.click()
    row = [tr.text for tr in td]
    row = row[1:11]+row_detail[0:8]
    results_2019.append(row)
    print(f'Runner {runner_id}/{len(table_2019_rows)}')

results_df_2019 = pd.DataFrame(results_2019, columns=["overall_place",
                                                     "bib",
                                                     "name",
                                                     "birth_year",
                                                    "nationality",
                                                    "club",
                                                    "gender_place",
                                                    "age_group_place",
                                                    "netto_time",
                                                    "brutto_time",
                                                    "5km",
                                                    "10km",
                                                    "15km",
                                                    "hm",
                                                    "25km",
                                                    "30km",
                                                    "35km",
                                                    "40km"])
```
